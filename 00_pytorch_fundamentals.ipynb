### Matrix multiplication

Two main ways of performing multiplication in neural networkd and deep learning:
1. Element-wise multiplication
2. Matrix multiplication (dot product)

There are two main rules that performing matrix multiplication need to satisfy:
1. The **inner dimensions** must mach:
* `(3,2) @ (2,3)` will work
* `(2,3) @ (3,2)` will work
* `(3,2) @ (3,2)` won't work
2. The resulting matrix has the shape of the **outer dimensions**:
* `(3,2) @ (2,3)` -> `(3, 3)`
* `(2,3) @ (3,2)` -> `(2, 2)`


```python
# Element wise multiplication
print(tensor, '*', tensor)
print(f"Equals: {tensor * tensor}")
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor)
```

### One of the most common errors in deep learning: shape errors

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])
tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])
```

```python
tensor_A.shape, tensor_B.shape
```

To fix out tensor shape issues, we can manipulate the shape of one of our tensors using a **transpose**
A **transpose** switches the axes or dimensions of a given tensor.


```python
tensor_B.T
```

```python
tensor_B
```

```python
tensor_B.shape, tensor_B.T.shape
```

```python
torch.mm(tensor_A, tensor_B.T) # torch.matmul = torch.mm
```

```python
torch.mm(tensor_A, tensor_B.T).shape
```

```python
# The matrix multiplication operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape}, tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

## Reshaping, stacking, squeezing and unsqueezing tensors
* Reshaping - reshapes an input tensor to a defined shape
* View - return a view of an input tensor of certain shape but keep the same momory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeeze - removes all `1` dimensions from a tensor
* Unsqueeze add a `1` dimension to a target tensor
* Permute - return a view of the input with dimensions permuted (swapped) in a certain way

```python
# Let's create a tensor
import torch
x = torch.arange(1., 11.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 10) # we just add a single dimension (input size = 10)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(10, 1)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(2, 5) # same dimension 2 x 5 as the input
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1,10)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor shares the same memory as the orighinal input)
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim = 0)
x_stacked
```

```python
# torch.squeeze() - remove all `1` dimensions from target input
x = torch.arange(1., 10)
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
print(f"Previous tensor: {x_reshaped}")
print(f"Previos shape: {x_reshaped.shape}")

# Remove extra dimesnion from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
# torch.unsqueeze() - adds a single dimension to a target tensor at a specific dim
print(f"Previous target: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Add an extra dimesion with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
# torch.permute - rearranges the dimensions of the target tensor to a specific order
x_original = torch.rand(size=(224, 224, 3)) # [height, width, colour_channels]

# Permute the original tensor to rearrange the azis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shift axis 0->2, 1->0, 2->1

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")   # [colour_channels, height, width]
                                          # x_permuted will take the same place in memory as x_original is just a view of the input tensor
```

## PyTorch tensors & NumPy

NumPy is a popular scientific Python numerical computing library.

And because of this, PyTorch has functionality to interact with it.

* Data in NumP, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* PyTorch tensor -> NumPy -> `torch.Tensor.numpy()`

```python
# NumPy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array) # warning: when converting from numpy to pytorch. Pytorch reflects numpy's default datatype of float 64 unless specified otherwise
array, tensor
```

```python
# Change the value of array, what will this do to 'tensor'?
array = array + 1
array, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# Change the tensor, what happens to 'numpy_tensor'?
tensor = tensor + 1
tensor, numpy_tensor
```

## Reproducibility (trying to take random out of random)

In short how a neural network learns:

`start with random numbers -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> again -> again...`

To reduce the randomness in neural networks and PyTorch come the concept of a **random seed**

Essentially what the random seed does is "flavour" the randomness.

```python
import torch

# Create two random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Let's make some random but reproductible tensors
import torch

# Set the random seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)      # Only work for one peace a code, we need to reset each time the random seed
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
